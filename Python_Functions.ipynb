5.) What are iterators in Python and how do they differ from iterables?
- In Python, an iterator is an object that produces the next value in a sequence each time you ask for it, while an iterable is any object you can loop over (like a list or string) that can give you an iterator
- In Python, iterables and iterators are closely related, but they are not the same thing
##### Iterable
- An iterable is an object that can be iterated over, meaning it can return its members one at a time. Examples include lists, tuples, strings, dictionaries, and range objects.
- < An object is considered iterable if it implements the __iter__() method, which returns an iterator, or the __getitem__() method which allows indexing >
      

      - my_list = [1, 2, 3]  # This is an iterable
        for item in my_list:
            print(item)

#### Iterator
- An iterator in Python is an object used to traverse through all the elements of a collection (like lists, tuples or dictionaries) one element at a time. It follows the iterator protocol, which involves two key methods
- < __iter__(): Returns the iterator object itself >
- < __next__(): Returns the next value from the sequence. Raises StopIteration when the sequence ends >


      - my_list = [1, 2, 3]
        my_iterator = iter(my_list)  # This creates an iterator from the iterable

        print(next(my_iterator))  # Output: 1
        print(next(my_iterator))  # Output: 2
        print(next(my_iterator))  # Output: 3
        # next(my_iterator) would raise StopIteration

6.) Explain the concept of generators in Python and how they are defined?
- A generator function is a special type of function that returns an iterator object. Instead of using return to send back a single value, generator functions use yield to produce a series of results over time
- This allows the function to generate values and pause its execution after each yield, maintaining its state between iterations
###### Generator are defined in two ways:-

A.) Generator Functions
- A generator function looks like a normal function but uses yield instead of return

        - def count_to_three():
              yield 1

              yield 3
          gen = count_to_three()
          print(next(gen))  # 1
          print(next(gen))  # 2

7.) What are the advantages of using generators over regular functions?
- Generators in Python have several advantages over regular functions
##### Characteristics and Benefits of Using Python Generators
- Efficient:-
Python Generators are more efficient than loops for large datasets, as they produce values one by one instead of storing them in memory before returning them.
- Memory Management:-
As generators produce values one by one rather than storing them in a list or other data structure, they also require less memory and can process large datasets without much RAM.
- Time-Saving:-
Generators in Python can also save time as they do not need to wait for the entire sequence to be generated before returning them, allowing code to execute in linear time.
- Infinite Sequences:-
Generators can produce infinite sequences, which is helpful for tasks such as stream processing or other activities requiring continuing input.

```python
# 1.) Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.

def sum_of_even_numbers(numbers):
    total = 0
    for num in numbers:
        if num % 2 == 0:  # check if the number is even
            total += num
    return total

nums = [1, 2, 3, 4, 5, 6]
print(sum_of_even_numbers(nums))
```

```python
# 4.)  Write a Python function that checks if a given number is prime or not from 1 to 200.

def is_prime(n):
  if n<2:    # 0 and 1 are not prime
    return False

  for i in range (2,n):
    if n % i == 0:   # if divisible by any number
      return False

  return True  # number is prime

for num in range (1, 201):
    if is_prime(num):
      print(num)


```

```python
# 5.) Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms

class Fibonacci:
  # now we make contructor
  def __init__(self, terms):
    self.terms = terms
    self.count = 0
    self.a = 0
    self.b = 1

    # for making iteraror we use __iter__
  def __iter__(self):
    return self  # object itself is an iterator

  def __next__(self):
    # for stoping the itteration if the required term is completed
    if self.count >= self.terms:
      raise StopIteration

    # for first term
    if self.count == 0:
      self.count += 1
      return self.a  # it return = 0

    # for second term
    if self.count == 1:
      self.count += 1
      return self.b  #it return = 1

    # for remaining terms
    next_value = self.a + self.b
    self.a = self.b
    self.b = next_value
    self.count += 1
    return next_value  # it return = next value

# now we use iterator
for num in Fibonacci(10):
  print(num)
```

```python
# 6.)  Write a generator function in Python that yields the powers of 2 up to a given exponent.

def power_of_two(n):
  for i in range (0, n+1): # from 0 to 1
    yield 2 ** i # here yiel generate square of i

for value in power_of_two(5):
  print(value)

```

```python
# 7.) Implement a generator function that reads a file line by line and yields each line as a string.

with open("file.txt", "w") as f: # "with open" open file in safe mode, and with w mode it overwrite a file
    f.write("This is line 1\n") # writing first line file and add new lime
    f.write("This is line 2\n")

def read_file(filename):
  file = open(filename, "r") # for openning file
  for line in file:  # for reading file line by line
    yield line  # yield retun line from file

  file.close()  # for closing file

for line in read_file("file.txt"):
  print(line)
```

```python
# 9.) Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

def celcius_to_fahrenheit(celcius):
  return (celcius * 9/5) + 32

celsius_temp = [0,25,32,46,10,20]
print("tem in celsius:", celsius_temp)

fahrenheit_temp = list(map(celcius_to_fahrenheit,celsius_temp))
print("tem in farenheit:",fahrenheit_temp)

```

```python
# 10.) Create a Python program that uses `filter()` to remove all the vowels from a given string.

def remove_vowels(input_string):
  vowels = "aeiouAEIOU"
  return input_string not in vowels

text = "HELLO how are you madam"

# filter send each character to the function and take only non vowels character
result = " " . join(filter(remove_vowels, text )) # here join is joining all character and making a string and return or save it to the result
print(result)

```
